```python
#dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#view data
usa_data = "resources/USvideos.csv"

usa_df = pd.read_csv(usa_data)
```

```python
usa_df.info()
```

```python
#drop unneccessary columns
yt_usa = usa_df.drop(columns=['video_id','trending_date','channel_title','publish_time',
'tags','thumbnail_link','comments_disabled','ratings_disabled','video_error_or_removed','description'])
```

```python
yt_usa.head()
```

```python
total_views = yt_usa['views'].count()
sum_views = yt_usa['views'].sum()
total_bycat = yt_usa.groupby("category_id")['views'].count()
```

```python
total_bycat
```

```python
music = 6472
entertainment = 9964
other = total_views - music
ent_other = total_views - entertainment
```

```python
ent_other
```

```python
labels = ["Entertainment","Other"]
sizes = [9964,30985]
colors = ["lightskyblue", "lightcoral"]
explode = (0.1,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%", shadow=False, startangle=100)
plt.axis("equal")
plt.title("Entertainment/Total Videos")
plt.show()
```
