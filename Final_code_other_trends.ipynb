```python
# Dependencies
#%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load in csv
youtube_df = pd.read_csv("Datasets_PlanB/USvideos_new_short.csv")
youtube_df.head()
```

```python
sum_views = youtube_df.groupby("category_desc")["views"].sum()
sum_views_sorted = sum_views.sort_values(ascending=False).head()
sum_views_sorted
```

```python
cats = ["Film & Animation", "Comedy", "People & Blogs", "Sports", "How-to & Style"]
views = [5388304721, 4268896024, 4075589569,4045001486,3369314433]

fig = plt.gcf()
fig.set_size_inches(10.5,5.5)
plt.bar(cats, views, align='center', color='b', alpha=0.75)
plt.title("YouTube Top 5 Video Categories")
plt.xlabel("Category Name")
plt.ylabel("# of Views (in Billions)")
plt.show()
```

```python
#Grouping by type in the effort to be able to use in all charts created
type_group = youtube_df.groupby('category_desc')['views', 'likes', 'dislikes', 'comment_count']

views_by_categories = type_group.sum()['views'].sort_values(ascending=False).head()

# Build Pie Chart
labels = views_by_categories.index
```

```python
#create color scheme that will be used for all charts and explode
colorsbycats = {'Sports': 'lightcoral', 'How-to & Style': 'lightskyblue', 'Film & Animation': 'gold', 'Comedy': 'magenta','People & Blogs': 'green'}
colors = [colorsbycats[n] for n in labels]
explode = [0, 0, 0, 0, 0]

#total views by categories
plt.pie(views_by_categories, startangle = 150, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = False)

#Final Pie Chart Displayed with Title and equal axis
plt.title('% of Views by Top 5 Categories')
plt.axis("equal")
plt.show()
```

```python
#total likes by categories
likes_by_categories = type_group.sum()['likes'].sort_values(ascending=False).head()

# Build Pie Chart
#labels = likes_by_categories.index
plt.pie(likes_by_categories, startangle = 150, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = False)

#Final Pie Chart Displayed with Title and equal axis
plt.title('% of Likes by Top 5 Categories')
plt.axis("equal")
plt.show()
```

```python
#total dislikes by categories
dislikes_by_categories = type_group.sum()['dislikes'].sort_values(ascending=False).head()

# Build Pie Chart
#labels = dislikes_by_categories.index
plt.pie(dislikes_by_categories, startangle = 150, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = False)

#Final Pie Chart Displayed with Title and equal axis
plt.title('% of Dislikes by Top 5 Categories')
plt.axis("equal")
plt.show()
```

```python
#total dislikes by categories
comments_by_categories = type_group.sum()['comment_count'].sort_values(ascending=False).head()

# Build Pie Chart
#labels = comments_by_categories.index
plt.pie(comments_by_categories, startangle = 150, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = False)

#Final Pie Chart Displayed with Title and equal axis
plt.title('% of Comments by Top 5 Categories')
plt.axis("equal")
plt.show()
```
